```python
# Example usage in a notebook or script
from synthetic_data.__dataset_direction_modifier import DatasetDirectionModifier
import pandas as pd
from datasets import load_dataset

# Load the dataset
dataset_name = "ruthchy/semantic-length-generalization-logo-image"
ds = load_dataset(dataset_name)

# Create modifier instance
modifier = DatasetDirectionModifier(random_seed=42)

# Modify the dataset
modified_ds, summary_df = modifier.replace_direction(
    dataset=ds,
    source_dir="left", 
    target_dir="right",
    proportion=0.5,
    return_overview=True
)
print(summary_df)
print("\n")
print("\n")

# Generate and display the summary
summary_df = modifier.modification_overview(ds, modified_ds, "left", "right")
print(summary_df)
```
